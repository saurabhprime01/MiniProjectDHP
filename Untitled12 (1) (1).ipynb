```python
import numpy as np
import pandas as pd
import csv

df = pd.read_csv("/Users/saura/OneDrive/Desktop/2nd Semester/Data Frame/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df
```

```python
# 1.	Calculate total revenue by contract type.
```

```python
total_revenue = df.groupby("Contract")["TotalCharges"].mean()
total_revenue
```

```python
# 2.	Find average monthly charges by contract type.
```

```python
avg_revenue = df.groupby("Contract")["MonthlyCharges"].sum()
avg_revenue 
```

```python
#Calculate churn rate.
```

```python
churn_rate = (df["Churn"].value_counts(normalize=True))*100
churn_rate
```

```python
# 4.Identify customers with tenure greater than 24 months.
```

```python
long_tenure = df[df["tenure"] > 24]
long_tenure[["customerID", "tenure"]]
```

```python
# 5.Find the top 10 highest paying customers (based on TotalCharges).
```

```python
top_10_customer = df.sort_values(by = "TotalCharges", ascending = False).head(10)
top_10_customer[["customerID", "TotalCharges"]]
```

```python
# 6.Compare average monthly charges between churn and non-churn customers.
```

```python
churn_avg = df.groupby("Churn")["MonthlyCharges"].mean()
churn_avg
```

```python
# 7.Find the contract type with the highest churn rate.
```

```python
churn_rate = df.groupby("Contract")["Churn"].value_counts(normalize = True)
churn_rate
```

```python
# 8.Identify customers with monthly charges greater than 80.
```

```python
high_charges = df[df["MonthlyCharges"]>80]
high_charges[["customerID", "MonthlyCharges"]]
```

```python
# 9.Calculate average tenure by contract type.
```

```python
avg_tenure = df.groupby("Contract")["tenure"].mean()
avg_tenure
```

```python
# 10.	Find total customers per contract type.
```

```python
total_customer = df["Contract"].value_counts()
total_customer
```
